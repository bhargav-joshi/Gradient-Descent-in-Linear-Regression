## Gradient Descent 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

**Why This ?**
Given a Function f(x), we want to **find the value of x that minimizes f.**


## Visualisation

```python
plt.plot(X,Y)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

```python
x = 0
lr = 0.1
error = []

plt.plot(X,Y,color='Green')

for i in range(50):
    grad = 2*(x-5)
    x = x - lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
# Plot the values of error

plt.plot(error,color='red')
plt.show()
```

## Surface Plots | Data Visualisation

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(-1,1,0.02)
b = a 
a,b = np.meshgrid(a,b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.title("Contour Plot")
plt.show()
```
